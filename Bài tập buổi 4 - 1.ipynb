```python
import random

def choi_game():
    print("Chào mừng bạn đến với trò chơi Đoán Số!")
    so_lan_thang = 0
    so_lan_thua = 0

    while True:
        print("\nChọn cấp độ chơi:")
        print("1. Dễ (9 lần đoán)")
        print("2. Vừa (6 lần đoán)")
        print("3. Khó (4 lần đoán)")
        
        while True:
            cap_do = input("Nhập cấp độ (1-3): ")
            if cap_do in ["1", "2", "3"]:
                break
            print("Vui lòng nhập 1, 2 hoặc 3.")

        if cap_do == "1":
            so_lan_doan = 9
        elif cap_do == "2":
            so_lan_doan = 6
        else:
            so_lan_doan = 4

        so_bi_mat = random.randint(1, 100)
        print(f"\nTôi đã chọn 1 số từ 1 đến 100. Bạn có {so_lan_doan} lần để đoán.")

        for lan in range(1, so_lan_doan + 1):
            while True:
                try:
                    doan = int(input(f"Lần {lan}: Nhập số bạn đoán: "))
                    if 1 <= doan <= 100:
                        break
                    else:
                        print("Vui lòng nhập số từ 1 đến 100.")
                except ValueError:
                    print("Vui lòng nhập số hợp lệ.")

            if doan < so_bi_mat:
                print("Số bạn đoán nhỏ hơn.")
            elif doan > so_bi_mat:
                print("Số bạn đoán lớn hơn.")
            else:
                print("Chúc mừng! Bạn đã đoán đúng!")
                so_lan_thang += 1
                break
        else:
            print(f"Bạn đã hết lượt đoán. Số đúng là: {so_bi_mat}")
            so_lan_thua += 1

        choi_tiep = input("Bạn có muốn chơi tiếp không? (c/k): ").lower()
        if choi_tiep != 'c':
            print("\nTrò chơi kết thúc.")
            print(f"Tổng số lần thắng: {so_lan_thang}")
            print(f"Tổng số lần thua : {so_lan_thua}")
            break

choi_game()
```
